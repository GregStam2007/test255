```python
cap = cv.VideoCapture(0)

if not cap.isOpened():
    print("Î”ÎµÎ½ Î²ÏÎ­Î¸Î·ÎºÎµ ÎºÎ¬Î¼ÎµÏÎ±.")
    exit()

try:
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        frame = cv.resize(frame, (720, 860))

        # Conversion to  HSV
        hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)

        # Split channels
        h, s, v = cv.split(hsv)

        # Creation of colored versions for better visualization
        h_vis = cv.applyColorMap(h, cv.COLORMAP_JET)
        s_vis = cv.applyColorMap(s, cv.COLORMAP_BONE)
        v_vis = cv.applyColorMap(v, cv.COLORMAP_HOT)

        # Combination of channels (horizontally)
        top = np.hstack((h_vis, s_vis, v_vis))

        frame_large = cv.resize(frame, (top.shape[1], top.shape[0]))


        all_view = np.vstack((frame_large, top))

        clear_output(wait=True)
        display(Image.fromarray(cv.cvtColor(all_view, cv.COLOR_BGR2RGB)))

        time.sleep(0.01)

except KeyboardInterrupt:
    print("Î¤ÎµÏÎ¼Î±Ï„Î¹ÏƒÎ¼ÏŒÏ‚ Î±Ï€ÏŒ Ï‡ÏÎ®ÏƒÏ„Î·.")

cap.release()
```
